```python
!!!第一題!!!

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T
target = iris.target
species = iris.target_names
#colors = ['Wheat', 'PaleGreen', 'PaleTurquoise']PaleGoldenRodWheatPaleTurquoisePapayaWhipSkyBlue
colors = ['Wheat', 'PaleGreen', 'PaleTurquoise']

mean_values = np.zeros((3, 4))
std_values = np.zeros((3, 4))

# Calculate mean and standard deviation
for k in range(3):
    species_data = features[:, target == k]
    mean_values[k, :] = np.mean(species_data, axis=1)
    std_values[k, :] = np.std(species_data, axis=1)

for i in range(3):
    print(f"Species: {species[i]}")
    for j in range(4):
        print(f"  Feature {iris.feature_names[j]}: Mean = {mean_values[i, j]:.2f}, Std = {std_values[i, j]:.2f}")

# Create scatter plots
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for i in range(4):
    ax = axes[i]
    for k in range(3):
        ax.scatter(mean_values[k, i], std_values[k, i], alpha=0.8, s=200, c=colors[k], label=f'Species {species[k]}')
    ax.set_xlabel('Mean')
    ax.set_ylabel('Standard Deviation')
    ax.set_title(f'{iris.feature_names[i]}')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
!!!第二題!!!

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T
target = iris.target
species = iris.target_names
colors = plt.cm.Set3(np.linspace(0, 1, 3)) # Using tab10 colormap with 3 distinct colors

fig, axes = plt.subplots(4, 4, figsize=(15, 15))

for i in range(4):
    for j in range(4):
        ax = axes[i, j]
        if i == j:  # Histogram
            for k in range(3):
                x = features[i][target == k]
                ax.hist(x, alpha=0.3, bins=20, label=species[k])
            ax.legend()
        else:  # Scatter plot
            for k in range(3):
                x = features[i][target == k]
                y = features[j][target == k]
                ax.scatter(x, y, alpha=0.5, s=50, c=colors[k], label=species[k])
                #ax.scatter(x, y, alpha=0.2, s=50, c=target[target == k], cmap='viridis_r', label=species[k])
            ax.legend()
        ax.set_xlabel(iris.feature_names[i])
        ax.set_ylabel(iris.feature_names[j])
        ax.set_title(f'{iris.feature_names[i]} vs {iris.feature_names[j]}')

plt.tight_layout()
plt.show()
```
